```python
def sum_sub_abs(arr1, arr2, cut_rate=0.25, avg_rate=0.25):
    arr = arr1 - arr2
    arr = np.where(abs(arr)<cut_rate, 0, arr)
    arr_avg = abs(arr1+arr2)/2 
    arr_avg = np.where(arr_avg<avg_rate, 1, arr_avg)
    _v, _w = arr.shape
    _sum = np.sum(abs(arr))/(_v*_w)
    return _sum

def cal_abs_diff(arr1, arr2):
    _v1, _w1 = arr1.shape
    _v2, _w2 = arr2.shape
    arr1 = arr1/np.max(abs(arr1))
    arr2 = arr2/np.max(abs(arr2))
    if _v1 != _v2:
        print('vertical of cal_dhash_diff_rate(arr1, arr2) must be same')
        return []
    if _w1 == _w2:
        _out = sum_sub_abs(arr1, arr2)
        return _out
    if _w1 > _w2:
        _d = _w1 - _w2
        t_val = np.zeros(_d+1)
        for k in range(_d+1):
            t_val[k] = sum_sub_abs(arr1[:, k:k+_w2], arr2[:, 0:_w2])
        return np.min(t_val)
    else:
        _d = _w2 - _w1
        t_val = np.zeros(_d+1)
        for k in range(_d+1):
            t_val[k] = sum_sub_abs(arr1[:, 0:_w1], arr2[:, k:k+_w1])
        return np.min(t_val)  
    
abs_arr = np.zeros((12,12))
abs_arr_2 = np.zeros((12,12))
for i in range(12):
    for j in range(12):
        _wi = pars_pc['arr_0'][i]['_zoom_width']
        _wj = pars_01['arr_0'][j]['_zoom_width']
        abs_arr[i,j] = cal_abs_diff(pars_pc['arr_0'][i]['spec_zoom_out'][:,0:_wi//2],
                                    pars_01['arr_0'][j]['spec_zoom_out'][:,0:_wj//2])
        
        abs_arr_2[i,j] = cal_abs_diff(pars_pc['arr_0'][i]['spec_zoom_out'][:,0:_wi],
                                    pars_01['arr_0'][j]['spec_zoom_out'][:,0:_wj])
                
plt.imshow(abs_arr*10)
plt.show()
print(abs_arr*10)
print(np.min(abs_arr*10))
print(np.max(abs_arr*10))

plt.imshow(abs_arr_2*10)
plt.show()
print(abs_arr_2*10)
print(np.min(abs_arr_2*10))
print(np.max(abs_arr_2*10))
```

```python
def subtract_dhash(arr1, arr2):
    arr = arr1*arr2
    _v, _w = arr.shape
    arr = np.where(arr < 0, 1, 0)
    _sum = np.sum(arr*10)/(_v*_w)
    return _sum

def sub_abs_dhash(arr1, arr2):
    arr = arr1 - arr2
    _v, _w = arr.shape
    _sum = np.sum(abs(arr))/(_v*_w)
    return _sum

def cal_dhash_diff_rate(arr1, arr2):
    _v1, _w1 = arr1.shape
    _v2, _w2 = arr2.shape
    if _v1 != _v2:
        print('vertical of cal_dhash_diff_rate(arr1, arr2) must be same')
        return []
    if _w1 == _w2:
        _out = sub_abs_dhash(arr1, arr2)
        return _out
    if _w1 > _w2:
        _d = _w1 - _w2
        t_val = np.zeros(_d+1)
        for k in range(_d+1):
            t_val[k] = sub_abs_dhash(arr1[:, k:k+_w2], arr2[:, 0:_w2])
        return np.min(t_val)
    else:
        _d = _w2 - _w1
        t_val = np.zeros(_d+1)
        for k in range(_d+1):
            t_val[k] = sub_abs_dhash(arr1[:, 0:_w1], arr2[:, k:k+_w1])
        return np.min(t_val)  

simi_arr = np.zeros((12,12))
for i in range(12):
    for j in range(12):
        _wi = pars_pc['arr_0'][i]['_zoom_width']
        _wj = pars_01['arr_0'][j]['_zoom_width']
        simi_arr[i,j] = cal_dhash_diff_rate(pars_pc['arr_0'][i]['spec_dhash'][:,0:_wi],
                                           pars_01['arr_0'][j]['spec_dhash'][:,0:_wj])
                
plt.imshow(simi_arr)
plt.show()

print(simi_arr)
```
