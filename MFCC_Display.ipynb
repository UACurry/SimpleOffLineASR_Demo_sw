```python
wav_file = "./ResFiles/Wav/Word_voice_tmpt.wav"
wav, sr = librosa.load(wav_file)
print("sr: ", sr)
print("len(wav):",len(wav))
print(wav.dtype)
```

```python
import wave 
wav_obj = wave.open(wav_file)
print(wav_obj)
samples = wav_obj.getnframes()
print(samples)
paras = wav_obj.getparams()
print(paras)
data = wav_obj.readframes(samples)
wav_obj.close()
wav_data = np.frombuffer(data, dtype=np.int16)
```

```python
plt.plot(wav_data)
plt.grid(True)
plt.title("语音波形曲线")
plt.show()
```

```python
f_wav = np.array(wav_data, dtype=np.float32)
S = librosa.feature.melspectrogram(y=f_wav, sr=16000, n_mels=128,fmax=1000)
mfcc = librosa.feature.mfcc(S=librosa.power_to_db(S))
# mfcc = librosa.feature.mfcc(S)
# print(mfcc.shape)

plt.figure(figsize=(10,8))
plt.imshow(mfcc[2:,:],cmap="jet", extent=(0,433,0,100),origin="lower")
plt.title("MFCC")
plt.show()
```

```python
%timeit S = librosa.feature.spectral_centroid(y=f_wav, sr=16000)
%timeit S = librosa.feature.melspectrogram(y=f_wav, sr=16000, n_mels=128,fmax=1000)
%timeit mfcc = librosa.feature.mfcc(S=librosa.power_to_db(S))
```

```python
curves_idx = np.array((25,50,55, 75,80,115,120, 140,145, 170,175,200,205))
curve_mfcc = np.zeros((len(curves_idx), 20))
for i in range(len(curves_idx)):
    curve_mfcc[i,:] = mfcc[:,curves_idx[i]].T

plt.figure(figsize=(10,8))
for i in range(len(curves_idx)):
    plt.plot(curve_mfcc[i,1:], '-x', label=str(i)) 
plt.grid(True)
plt.legend()
plt.show()
```

```python
y = f_wav
sr = 16000
m_slaney = librosa.feature.mfcc(y=y, sr=sr, dct_type=2)
m_htk = librosa.feature.mfcc(y=y, sr=sr, dct_type=3)
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
librosa.display.specshow(m_slaney, x_axis='time')
plt.title('RASTAMAT / Auditory toolbox (dct_type=2)')
plt.colorbar()
plt.subplot(2, 1, 2)
librosa.display.specshow(m_htk, x_axis='time')
plt.title('HTK-style (dct_type=3)')
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
y = f_wav
zcr = librosa.feature.zero_crossing_rate(y,
                    frame_length = 512, 
                    hop_length = 256, center = True)

stft = librosa.stft(y, n_fft=512, hop_length=None, 
             win_length=None, window='hann', 
             center=True, pad_mode='reflect')

print(stft.shape)

plt.figure(figsize=(10,10))
plt.subplot(311)
plt.plot(wav_data)
plt.grid(True)
plt.title("语音波形曲线")
plt.subplot(312)
plt.plot(zcr.T)
plt.grid(True)
plt.title("过零率")
plt.subplot(313)
plt.imshow(np.abs(stft)[0:50,:], extent=(0,1729,0,500),origin="lower")
# plt.grid(True)
plt.title("stft")

plt.show()
```

```python
# https://www.cnblogs.com/LXP-Never/p/11561355.html
S = np.abs(librosa.stft(y))
# print(librosa.power_to_db(S ** 2))

plt.figure(figsize=(10,10))
plt.subplot(2, 1, 1)
# 从波形获取功率谱图
librosa.display.specshow(S ** 2, sr=sr, y_axis='log')  

plt.colorbar()
plt.title('功率谱 Power spectrogram')
plt.subplot(2, 1, 2)
# 相对于峰值功率计算dB, 那么其他的dB都是负的，注意看后边cmp值
librosa.display.specshow(librosa.power_to_db(S ** 2, ref=np.max),
                         sr=sr, y_axis='log', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('dB功率谱 Log-Power spectrogram')
# plt.set_cmap("autumn")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,10))
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
plt.subplot(2, 1, 1)
librosa.display.specshow(D, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('线性频率功率谱')

plt.subplot(2, 1, 2)
librosa.display.specshow(D, y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('对数频率功率谱')
plt.show()
```

```python
# # 方法一：使用时间序列求Mel频谱
# print(librosa.feature.melspectrogram(y=y, sr=sr))
# 方法二：使用stft频谱求Mel频谱
D = np.abs(librosa.stft(y)) ** 2  # stft频谱
S = librosa.feature.melspectrogram(S=D)  # 使用stft频谱求Mel频谱

plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(S, ref=np.max),
                         y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
plt.show()
```

```python
# 提取 mel spectrogram feature
melspec = librosa.feature.melspectrogram(y, sr, n_fft=1024, hop_length=512, n_mels=128)
logmelspec = librosa.amplitude_to_db(melspec)        # 转换到对数刻度

# print(logmelspec.shape)

plt.figure(figsize=(10, 4))
logmelspec = logmelspec - np.min(logmelspec)
plt.imshow(logmelspec,extent=(0,433,0,200),cmap="jet", origin="lower")
plt.colorbar()
plt.title("Log-Mel Spectrogram")
plt.show()
```

```python
# 提取 MFCC feature
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)

# print(mfccs.shape)   

plt.figure(figsize=(10, 4))
logmelspec = logmelspec - np.min(logmelspec)
plt.imshow(mfccs[2:, :],extent=(0,433,0,200),cmap="jet", origin="lower")
plt.colorbar()
plt.title("MFCC feature")
plt.show()
```
