```python
def subtract_dhash(arr1, arr2):
    arr = arr1*arr2
    _v, _w = arr.shape
    arr = np.where(arr < 0, 1, 0)
    _sum = np.sum(arr*10)/(_v*_w)
    return _sum

def sub_abs_dhash(arr1, arr2):
    arr = arr1 - arr2
    _v, _w = arr.shape
    _sum = np.sum(abs(arr))/(_v*_w)
    return _sum

def cal_dhash_diff_rate(arr1, arr2):
    _v1, _w1 = arr1.shape
    _v2, _w2 = arr2.shape
    if _v1 != _v2:
        print('vertical of cal_dhash_diff_rate(arr1, arr2) must be same')
        return []
    if _w1 == _w2:
        _out = sub_abs_dhash(arr1, arr2)
        return _out
    if _w1 > _w2:
        _d = _w1 - _w2
        t_val = np.zeros(_d+1)
        for k in range(_d+1):
            t_val[k] = sub_abs_dhash(arr1[:, k:k+_w2], arr2[:, 0:_w2])
        return np.min(t_val)
    else:
        _d = _w2 - _w1
        t_val = np.zeros(_d+1)
        for k in range(_d+1):
            t_val[k] = sub_abs_dhash(arr1[:, 0:_w1], arr2[:, k:k+_w1])
        return np.min(t_val)     
            

simi_arr = np.zeros((12,12))
for i in range(12):
    for j in range(12):
        _wi = pars_pc['arr_0'][i]['_zoom_width']
        _wj = pars_pc['arr_0'][j]['_zoom_width']
        simi_arr[i,j] = cal_dhash_diff_rate(pars_pc['arr_0'][i]['spec_dhash'][:,0:_wi],
                                           pars_pc['arr_0'][j]['spec_dhash'][:,0:_wj])
                
plt.imshow(simi_arr)
plt.show()

print(simi_arr)
```

```python
fil_3x3 = np.array([[1/18, 1/9, 1/18],
                   [1/9, 1/3, 1/9],
                   [1/18, 1/9, 1/18]])
# print(fil_3x3.shape)
# print(np.sum(fil_3x3))

def fil_33(arr, kernel):
    _v, _w = arr.shape
    
    _out_ext = np.zeros((_v+2, _w+2))
    
    _out_ext[1:_v+1, 1:_w+1] = arr
    
    _out_ext[0,1:_w+1] = arr[0,:]
    _out_ext[_v+1,1:_w+1] = arr[_v-1,:]
    _out_ext[1:_v+1,0] = arr[:,0]
    _out_ext[1:_v+1,_w+1] = arr[:,_w-1]
    
    _out = np.zeros((_v, _w))
    for i in range(_v):
        for j in range(_w):
            roi = _out_ext[i:i+3, j:j+3]
            mul = roi * kernel
            _out[i,j] = np.sum(mul)
    return _out

def find_peak(arr, cut_rate=0.1):
#     print(arr.shape)
    _v, _w = arr.shape
#     _out_ext = np.zeros((_v+2, _w+2))
#     _out_ext[1:_v+1, 1:_w+1] = arr
    _out = np.zeros((_v, _w))
    for i in range(1,_v-1):
        mask = (arr[i,:]>=arr[i-1,:]) & (arr[i,:]>=arr[i+1,:])
        _out[i,mask] = arr[i,mask]
#         print(mask.shape)
    return _out

spec_peak = np.zeros((12,80,80))
for i in range(12):
    _spec_fil = fil_33(pars_pc['arr_0'][i]['specgram'], fil_3x3)
    _spec_fil_2 = find_peak(_spec_fil)
    spec_peak[i] = fil_33(_spec_fil_2, fil_3x3)
    
plt.figure(figsize=(10,10))
for i in range(3):
    for j in range(4):
        plt.subplot(3,4,i*4 + j+1)
        plt.imshow(spec_peak[i*4 + j], origin='lower')
        plt.title(pars_pc['arr_0'][i*4 + j]['word'])
plt.show()
```

```python
def subtract_zoom_out(arr1, arr2, cut_rate=0.20):
    arr = arr1 - arr2
    _v, _w = arr.shape
    _rate = np.max(abs(arr))*cut_rate
    arr = np.where(abs(arr) >= _rate, arr, 0)
    _sum = np.sum(abs(arr))/(_v*_w)
    return _sum

simi_arr = np.zeros((12,12))
for i in range(12):
    for j in range(12):
        _wi = pars_pc['arr_0'][i]['_zoom_width']
        _wj = pars_pc['arr_0'][j]['_zoom_width']
        _w = min(_wi, _wj)
#         print(_w)
        simi_arr[i,j]=subtract_zoom_out(pars_pc['arr_0'][i]['spec_zoom_out'][0:10, 0:5],
                                       pars_pc['arr_0'][j]['spec_zoom_out'][0:10, 0:5])
plt.imshow(simi_arr)
plt.show()

np.set_printoptions(precision=2)
print(simi_arr)
```
